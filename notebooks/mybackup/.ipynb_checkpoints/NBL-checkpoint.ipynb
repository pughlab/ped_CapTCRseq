```r
flelst <- list.files(datapath, recursive = TRUE, pattern = "CLONES_TRB")

flelst <- flelst[grepl("Normal", flelst)]   
```

```r
mystatslist <- list.files(datapath,recursive = TRUE,pattern = "divstats") 

# only specified chain  
mystatslist <- mystatslist[grepl("TRB", mystatslist)]
 # only PBMC
mystatslist <- mystatslist[grepl("Normal", mystatslist)]
    
all.statsfiles <- lapply(mystatslist, function(f){ read.csv(paste0(datapath,f), header = TRUE, stringsAsFactors = F)})

div_trb_normal <- as.data.frame(do.call(rbind, all.statsfiles))

# Only CHP samples
div_trb_normal <- div_trb_normal[grepl("CHP", div_trb_normal$X),]

div_trb_normal$Patient <- gsub("_PBMC.*", "", div_trb_normal$X)
```

```r
div_trb_normal <- div_trb_normal[order(div_trb_normal$observed_Shannon),]  
    
 i <- 1
  for (f in flelst){
    mixcrfle <- read.table(paste(datapath, f, sep = ""), 
                           header = TRUE, sep = "\t",
                           stringsAsFactors = FALSE,
                           na.strings = c("", "NA"))

    if(i == 1){
      compldfle <- mixcrfle
      compldfle <- cbind(cloneno = row.names(compldfle), filename = f, compldfle)
      i <- i + 1   
    }
    else{
      compldfle1 <- mixcrfle
      compldfle1 <- cbind(cloneno = row.names(compldfle1), 
                          filename = f, compldfle1)
      compldfle <- rbind(compldfle, compldfle1)
      rm(compldfle1)
    }
  }

 compldfle$filename <- as.character(compldfle$filename)
  compldfle$cloneno <-  as.character(compldfle$cloneno)
  compldfle$Patient <- gsub(".*CLONES_TRB", "", compldfle$filename)
  compldfle$Patient <- gsub("_PBMC_DNA_2000000.txt", "",compldfle$Patient)
  #compldfle$Patient <- substr(compldfle$samplename, 5,7)
  compldfle$cloneno <- formatC(compldfle$cloneno, 
                               width=nchar(max(as.numeric(compldfle$cloneno))), flag="0")   

compldfle$Patient <- factor(compldfle$Patient, levels = div_trb_normal$Patient)  
    
compldfle <- compldfle[!is.na(compldfle$Patient),]
```

```r
  clonenocol = rep("#d9d9d9", 
                     nlevels(as.factor(compldfle$cloneno)))
  
  barpt <- ggplot(data = compldfle, 
                  aes(y = cloneCount, 
                      x = Patient, 
                      fill = cloneno)) + 
    geom_bar(colour = "#000000", stat="identity", width = 0.8) +
    theme(axis.title.x = element_blank(),
          axis.text.x = element_text(angle = 45, hjust = 1),
          axis.line = element_line(color = "black"),
          axis.text = element_text(size = 50),
          axis.title = element_text(size = 50), 
          legend.position = "none") +
    theme(panel.grid.major = element_blank(),
          panel.grid.minor = element_blank(),
          panel.background = element_rect(fill = "transparent",
                                          colour = NA),
          panel.border=element_blank(),
          plot.margin = unit(c(0,0,0,0),"cm")) +
    scale_fill_manual(values = clonenocol,
                      guide = FALSE) 
```
